```python
from stable_baselines3 import PPO
from stable_baselines3.ppo.policies import MlpPolicy
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.env_checker import check_env
import gym
import numpy as np
import sys
sys.path.append("../../")
from Hack import load, rl
%matplotlib qt5
```

```python
epex = load.epex().load()
price_array = epex['apx_da_hourly'].values
max_time = 30769 
max_time = 30769#2*24*28
env = rl.energy_price_env(price_array, max_time=max_time, window_size=24*2)
model = PPO(MlpPolicy, env, verbose=1)
check_env(env, warn=True)
```

```python
mean_reward_before_train = rl.evaluate(model, num_episodes=10, index = epex.index)
```

```python
# Train the agent for 10000 steps
model.learn(total_timesteps=10000)
model.save("../model_dir/trained_model")

# load model using loaded_model = PPO.load("path_to_model")
```

```python
# Trained Agent, after training
periods = 48*7
new_env =  DummyVecEnv([lambda: rl.energy_price_env(price_array, start_time=max_time, max_time = periods)])
mean_reward_after_train = rl.evaluate(model, new_env=new_env, num_episodes=10, index=epex.index)
```
